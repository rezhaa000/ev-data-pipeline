```python
import great_expectations as ge
import pandas as pd
from great_expectations.data_context import FileDataContext
```

```python
df = pd.read_csv('P2M3_Rezha_Aulia_data_clean.csv')

print(df.columns.tolist())
print(f'\nShape: {df.shape}')
print(f'\nSample Data:')
df.head()
```

```python
# Data Context
context = FileDataContext.create(project_root_dir='./')
```

```python
# Data Source
datasource_name = 'ev-data'
datasource = context.sources.add_pandas(datasource_name)

# Data asset
asset_name = 'ev-clean-data'
path_to_data = 'P2M3_Rezha_Aulia_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Batch Request
batch_request = asset.build_batch_request()
```

```python
# Expectation suite
expectation_suite_name = 'expectation-ev-data'
context.add_or_update_expectation_suite(expectation_suite_name)

# Validator
validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name=expectation_suite_name
)

validator.head()
```

## Expectation 1 : to_be_unique

```python
# Expectation 1 : dol_vehicle_id harus unik 
# DOL Vehicle ID adalah identifikasi unik setiap kendaraan yang diterbitkan oleh Washington State
validator.expect_column_values_to_be_unique(column='dol_vehicle_id')
```

## Expectation 2 : to_be_between

```python
# Expectation 2 : Electric Vehicle Range harus antara 0 dan 337 miles
# Jarak yang bisa ditempuh baterai EV tidak mungkin negatif dan juga tidak melebihi jarak max dataset
validator.expect_column_values_to_be_between(
    column='electric_range',
    min_value=0,
    max_value=337
)
```

## Expectation 3 : to_be_in_set

```python
# Expectation 3 : electric_vehicle_type hanya ada BEV atau PHEV
# Hanya ada 2 tipe kendaraan listrik yang terdaftar di Washington State
validator.expect_column_values_to_be_in_set(
    column='electric_vehicle_type',
    value_set=[
        'Battery Electric Vehicle (BEV)',
        'Plug-in Hybrid Electric Vehicle (PHEV)'
    ]
)
```

## Expectation 4 : to_be_in_type_list

```python
# Expectation 4 : electric_range harus bertipe numerik
# Kolom electric_range harus berupa angka untuk keperluan analisis dan perhitungan
validator.expect_column_values_to_be_in_type_list(
    column='electric_range',
    type_list=['int64', 'float64', 'int', 'float']
)
```

## Expectation 5 : expect_column_value_lenghts_to_be_between

```python
# Expectation 5 : Panjang karakter kolom vin harus antara 5 dan 10
# Kolom VIN (1-10) berisi 10 karakter pertama dari Vehicle Identification Number
# Panjang karakternya tidak boleh kurang dari 5 atau lebih dari 10

validator.expect_column_value_lengths_to_be_between(
    column='vin',
    min_value=5,
    max_value=10
)
```

## Expectation 6 : expect_column_values_to_match_regex

```python
# Expectation 6 : State harus berisi 2 huruf kapital
# Setelah filter hanya Washington State (WA), semua nilai state harusnya kode negara bagian
# 2 huruf kapital adalah formatnya USPS

validator.expect_column_values_to_match_regex(
    column='state',
    regex='^[A-Z]{2}$'
)
```

## Expectation 7 : expect_column_proportion_of_unique_values_to_be_between

```python
# Expectation 7 : Proporsi nilai unik kolom 'make' harus antara 0.0001 dan 0.01
# Kolom make berisi nama merek kendaraan dengan jumlah yang terbatas
validator.expect_column_proportion_of_unique_values_to_be_between(
    column='make',
    min_value=0.0001,
    max_value=0.01
)
```

```python
validator.save_expectation_suite(discard_failed_expectations=False)
print('Expectation suite berhasil disimpan')
```
